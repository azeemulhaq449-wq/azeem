```python
### 3) Simple Interest

P = float(input("Principal amount: "))
R = float(input("Rate of interest (%): "))
T = float(input("Time enter (saalon): "))

SI = (P * R * T) / 100

print("Simple Interest:", SI)
```

```python
### 4) Speed of an Object:
distance = float(input("Distance (km): "))

time = float(input("Time enter (hours): "))

speed = distance / time

print("Speed:", speed, "km/h")
```

```python
### 5) BMI Calculator: 

weight = float(input("Wazan (kg): "))
height = float(input("Height (meters ): "))

bmi = weight / (height ** 2)

print("Aapka BMI:", bmi)
```

```python
### 6) Force Using Newton's Second Law:
m = float(input("Mass dijiye (kg mein): "))
a = float(input("Acceleration dijiye (m/sÂ² mein): "))

force = m * a

print("Force hai:", force, "Newtons")
```

```python
###  7) Compound Interest:
P = float(input("Principal amount dijiye: "))
r = float(input("Annual interest rate (in decimal) dijiye: "))
n = int(input("Saale mein kitni baar compound hota hai (n): "))
t = float(input("Kitne saal (t): "))

A = P * (1 + r/n) ** (n*t)
CI = A - P

print("Compound Interest hai:", CI)
```

```python
### 8) Perimeter of a Triangle: 
a = float(input("Side a: "))
b = float(input("Side b: "))
c = float(input("Side c: "))

perimeter = a + b + c

print("Perimeter of triangle:", perimeter)
```

```python
### 10) Kinetic Energy:
m = float(input("Mass dijiye (kg mein): "))
v = float(input("Velocity dijiye (m/s mein): "))

ke = 0.5 * m * (v ** 2)

print("Kinetic Energy hai:", ke, "Joules")
```
